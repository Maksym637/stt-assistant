### Data generation using the `gTTS` library

````python
from gtts import gTTS
from IPython.display import Audio
````

````python
def generate_audio(text, filename="output.mp3", lang="en"):
    try:

        tts = gTTS(text=text, lang=lang, slow=False)
        tts.save(filename)

    except Exception as exc:
        print(f"Error '{exc}' - occured")
````

- - -

First phrase for voiceover:

```text
I enjoy studying foreign languages
```

````python
PHRASE_1 = {
    "en": "I enjoy studying foreign languages",
    "de": "Ich lerne gerne Fremdsprachen",
    "es": "Me gusta estudiar idiomas extranjeros",
}

PHRASE_1_PATH = {
    "en": "./tests/data/first-phrase-en.mp3",
    "de": "./tests/data/first-phrase-de.mp3",
    "es": "./tests/data/first-phrase-es.mp3",
}
````

````python
for lang, phrase in PHRASE_1.items():
    generate_audio(text=phrase, filename=f"first-phrase-{lang}.mp3", lang=lang)
````

````python
Audio(PHRASE_1_PATH["en"])
````

````python
Audio(PHRASE_1_PATH["de"])
````

````python
Audio(PHRASE_1_PATH["es"])
````
